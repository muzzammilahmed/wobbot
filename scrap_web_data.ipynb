```python

# Function to extract URLs from the sitemap XML
def extract_urls_from_sitemap(url):
    response = requests.get(url)  # Fetch the XML file
    if response.status_code == 200:
        # Parse the XML
        tree = ET.ElementTree(ET.fromstring(response.text))
        root = tree.getroot()

        # Namespaces dictionary to handle namespaces in the XML
        namespaces = {'': 'http://www.sitemaps.org/schemas/sitemap/0.9'}

        # Find all loc tags and extract URLs
        urls = []
        for loc in root.findall('.//url/loc', namespaces):
            urls.append(loc.text)

        return urls
    else:
        print("Failed to retrieve XML")
        return []

# Example usage:
xml_url = 'https://www.wob.ag/sitemap.site_1.xml'  # Replace this with your XML URL
urls = extract_urls_from_sitemap(xml_url)

with open("website_associated_links.json", "w", encoding="utf-8") as a_file:
    json.dump(urls, a_file)
# Print all extracted URLs
for url in urls:
    print(url)
```

```python

# Load URLs from the JSON file
with open("website_associated_links.json", "r", encoding="utf-8") as file:
    urls = json.load(file)

# Create the directory to store scraped data
output_folder = "text_data_scrapped"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Function to scrape and save data
def scrape_and_save(url):
    try:
        # Fetch the page content
        response = requests.get(url)
        if response.status_code == 200:
            # Parse the content using BeautifulSoup
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract some data, for example, the title and all paragraphs
            title = soup.title.string if soup.title else "No Title"
            paragraphs = soup.find_all('p')
            text_data = f"Title: {title}\n\n"
            
            # Concatenate all paragraph texts
            for paragraph in paragraphs:
                text_data += paragraph.get_text() + "\n"
            
            # Create a safe filename based on the URL
            file_name = url.replace("https://", "").replace("http://", "").replace("/", "_") + ".txt"
            file_path = os.path.join(output_folder, file_name)
            
            # Save the scraped data into a text file
            with open(file_path, "w", encoding="utf-8") as file:
                file.write(text_data)
            
            print(f"Scraped and saved data from {url}")
        else:
            print(f"Failed to retrieve {url}, Status Code: {response.status_code}")
    except Exception as e:
        print(f"Error while scraping {url}: {e}")

# Scrape data for each URL and save to individual text files
for url in urls:
    scrape_and_save(url)
```
